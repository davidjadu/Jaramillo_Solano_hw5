```python
import scikits.audiolab as audio
from __future__ import division
%pylab inline
```

```python
input_signal, sampling_rate, enc = audio.wavread("filosofia.wav")
time_array = arange(0, len(input_signal)/float(sampling_rate), 1/float(sampling_rate))
```

```python
plot(time_array,input_signal)
```

```python
from scipy.fftpack import fft, fftfreq
```

#Withe Noise
En el ruido blanco, el espectro de potencias tiene siempre la misma amplitud, luego la manera más fácil de hacer esto es dividir la transformada de fourier por su norma. Obteniéndose un espectro de potencias totalmente plano. Note que todas las fases se conservan

```python
fourier = fft(input_signal)
freq = fftfreq(len(time_array),time_array[1]-time_array[0])
```

```python
plot(freq, abs(fourier))
```

```python
fourier_withe_noise = fourier/abs(fourier)
```

```python
plot(freq, abs(fourier_withe_noise))
```

```python
white_noise_data = ifft(fourier_withe_noise)
```

```python
plot(white_noise_data.real)
```

```python
output_file = audio.Sndfile("withe_noise.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(white_noise_data.real)
output_file.close()
```

#Pink Noise
En el ruido rosado, el espectro de potencias es lineal en un gráfico logarítmico, luego la manera más fácil de hacer esto es dividir la transformada de fourier por su norma y posteriormente multiplicar por una exponencial decayente. Obteniéndose un espectro de potencias que se ve lineal en una gráfica logarítmica. Note que todas las fases se conservan.

```python
plot(freq, abs(fourier))
```

```python
maximum = freq[1:].max()
print maximum
fourier_pink = fourier/abs(fourier)*exp(-abs(freq)/maximum)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))
ax.set_yscale('log')
ax.plot(freq, abs(fourier_pink))
```

```python
pink_data = ifft(fourier_pink)
```

```python
plot(time_array, pink_data.real)
```

```python
output_file = audio.Sndfile("pink_noise.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(pink_data.real)
output_file.close()
```
