En este programa se busca hallar el radio de unos circulos que se encuentran en una imágen en la cual hay aparte de los circulos ruido.

```python
%pylab inline
from scipy.fftpack import fftfreq, fft
```

```python
data = np.loadtxt('circulos.dat')
x=data[:,0]
y=data[:,1]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

ax.scatter(x, y, alpha=0.5, s=1)
```

##Creación de la señal Bidimensional
Para generar una imagen completa a la cual se le pueda aplicar una función de correlación, se hizo lo siguiente, se creo una grilla de 1024x1024 (este tamaño para optimizar la transformada de fourier) y a cada punto de la grilla se le asignó la cantidad de puntos se encontraban dentro de ella. Esto se muestra a continuación.

```python
points = 1024
h=6.0/points
intensity=zeros((points,points))
for k in range(len(x)):
    intensity[int(y[k]/h), int(x[k]/h)]+=1.0
```

##Función de correlación
La función de correlación nos permite encontrar patrones repetitivos. Debido a que es el equivalente a suporponer la imágen consigo misma y empezar a moverla en todas las direcciones y en cada punto multiplicar su valor, entonces en los lugares donde los puntos se ajusten la correlación será grande, y donde no se ajusten será pequeña. Para más información ver John C. Russ, The image processing, 6 edition, pg. 391.
La función de autocorrelación se define (para funciones discretas, como la grilla generada anteriormente) como:

$$\rho_{x,y} = \frac{\sum_{i,j} f_{x+i,y+j}f_{i,j}}{\sqrt{\sum_{i,j}f_{x+i,y+j}^2 + \sum_{i,j}f_{i,j}^2  }}$$

donde el primer y segundo subíndice está modulo $i_{max}$, y modulo $j_{max}$ respectivamente

Ahora por el teorema de Wiener-Khinchin sabemos que la transormada de fourier de la autocorrelación es solo multiplicar la transformada de fourier de $f$ por su conjugada. Luego podemos obtener facilemnte esta transformada

```python
intensity_fft = fft2(intensity)
correlation_fft = fftshift(intensity_fft*intensity_fft.conjugate())
freq = fftshift(fftfreq(points,h))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im = ax.imshow(log(abs(correlation_fft)), origin='lower')
ax.set_title('log(Fourier transform)')
fig.colorbar(im, ax=ax)
```

##Correlación 1D
Como observamos en la figura, y como se espera ya que los patrones en la imagen son circulos que tienen simetría rotacional, observamos que aquí se tiene simetría rotacional, luego podemos definir una autocorrelación en una dimensión. para hacer esto tomamos círculos alrededor del centro y promediamos el valor de la autocorrelación.

```python
center = points/2
k1d=freq[freq>=0]
correlation1d_fft=zeros(points/2+1,dtype=np.dtype(np.complex128))
num=zeros(points/2+1)
for i in range(points):
    for j in range(points):
        k= int(sqrt((i-center)**2+(j-center)**2))
        if (k<=points/2):
            correlation1d_fft[k]+=correlation_fft[i,j]
            num[k]+=1
```

```python
correlation1d_fft=correlation1d_fft/num
```

```python
plot(log(abs(correlation1d_fft)))
```

#Selección de máximos
En la figura observamos que donde se encuentran más claros los máximos, son la zona de la izquierda, por tanto encontramos dichos máximos y la distancia entre cada dos máximos será el diámetro de la figura, debido a errores númericos, se tomaran todas las distancias y se promediarán, con este promedio se hallará el radio como:
$$
r=\frac{0.5}{\Delta k}
$$
Donde $\Delta k$ es el promedio de las distancias en unidades de frecuencia. el 0.5 es debido a que lo que se quiere es el radio, no el diámetro

```python
maxi = zeros(0)
values = zeros(0)
for i in range(5,150):
    actual = correlation1d_fft[i]
    if(actual> correlation1d_fft[i-1] and actual > correlation1d_fft[i+1]):
        values = append(values,[actual])
        maxi= append(maxi, i)
```
