```python
center=6.0*np.random.random((100,100))-3.0
R=0.31
theta = linspace(0,2*pi, 100)
x=zeros(0)
y=zeros(0)
```

```python
for c in center:
    x=append(x,[R*cos(theta)+c[0]])
    y=append(y,[R*sin(theta)+c[1]])
```

```python
scatter(x,y, alpha =0.5, s=2 )
```

```python
i=0
while(i<len(x)-1):
    if (abs(x[i])>3 or abs(y[i])>3):
        x=delete(x,i)
        y=delete(y,i)
        i-=1
    i+=1
```

```python
scatter(x,y, alpha =0.5, s=2 )
```

```python
points = 1024
h=6.0/points
intensity=zeros((points,points))
for k in range(len(x)):
    intensity[int(y[k]/h), int(x[k]/h)]+=1.0
```

```python
intensity_fft = fft2(intensity)
correlation_fft = fftshift(intensity_fft*intensity_fft.conjugate())
freq = fftshift(fftfreq(points,h))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im = ax.imshow(log(abs(correlation_fft)), origin='lower')
ax.set_title('log(Fourier transform)')
fig.colorbar(im, ax=ax)
```

```python
center = points/2
k1d=freq[freq>=0]
correlation1d_fft=zeros(points/2+1, dtype=np.dtype(np.complex128))
num=zeros(points/2+1)
for i in range(points):
    for j in range(points):
        k= int(sqrt((i-center)**2+(j-center)**2))
        if (k<=points/2):
            correlation1d_fft[k]+=correlation_fft[i,j]
            num[k]+=1
```

```python
correlation1d_fft=correlation1d_fft/num
```

```python
plot(log(abs(correlation1d_fft)))
```

```python
import operator
maximum,val= max(enumerate(abs(correlation1d_fft[160:200])), key=operator.itemgetter(1))
maximum+=160
```

```python
plot(log(abs(correlation1d_fft)))
scatter(maxi,log(values), color ='r', s= 2)
```
