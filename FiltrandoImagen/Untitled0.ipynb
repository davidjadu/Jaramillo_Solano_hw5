#Full Moon
full_moon.jpg

```python
img1 = plt.imread('full_moon.jpg')
dimy, dimx = shape(img1)[:2]
print dimx,dimy
```

```python
plt.imshow(img1)
```

```python
img1_fft = fft2(img1,axes=(0,1))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im=ax.imshow(abs(img1_fft))
ax.set_title('Fourier')
fig.colorbar(im, ax=ax)
```

```python
img_high_fft = img1_fft + 0.0
for i in range(dimy):
    for j in range(dimx):
        rad_squared = (i-150.0)**2+(j-148.0)**2
        if(rad_squared<20.0**2):
            img_high_fft[i,j,:]=[0,0,0]
        elif(rad_squared<30.0**2):
            img_high_fft[i,j,:]=img_high_fft[i,j,:]*sin(((sqrt(rad_squared)-20.0)/10.0)*pi/2.0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im=ax.imshow(abs(img_high_fft))
ax.set_title('High frquences fourier')
```
