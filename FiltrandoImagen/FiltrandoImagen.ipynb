```python
%pylab inline
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
```

#Analisis Imagen 1 


```python
Im1 = pylab.imread("colesterol-1.tif")
fig = figure(1, figsize=(10.5,8.5))
imshow(Im1, cmap = cm.Greys_r)
plt.title('Imagen Original', fontsize=12)

print shape(Im1)
```

```python
T1= fft2(Im1)
Tr1= fft2(Im1)
imshow(log(abs(fftshift(Tr1))))
colorbar()
```

#FiltroBajas
Solo se dejan pasar las frecuencias bajas

```python
low_freq = fftshift(Tr1)
ic=int(Tr1.shape[0])/2
jc=int(Tr1.shape[1])/2
a=5E-3
for i in range(Tr1.shape[0]):
    for j in range(Tr1.shape[1]):
        low_freq[i,j]*=exp(-a*((i-ic)**2 + (j-jc)**2))
```

```python
im1_low = ifft2(fftshift(low_freq))
imshow(abs(im1_low),  cmap=matplotlib.cm.Greys_r)
```

#FiltroAltas
Solo se dejan pasar las frecuencias altas

```python
high_freq = fftshift(Tr1)
ic=int(Tr1.shape[0])/2
jc=int(Tr1.shape[1])/2
a=5E-3
for i in range(Tr1.shape[0]):
    for j in range(Tr1.shape[1]):
        high_freq[i,j]*=(1-exp(-a*((i-ic)**2 + (j-jc)**2)))
```

```python
im1_high= ifft2(fftshift(high_freq))
imshow(abs(im1_high),  cmap=matplotlib.cm.Greys_r)
```
