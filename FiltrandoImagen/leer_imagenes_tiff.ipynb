#Análisis de imágenes en formato .tif
##Filtros Frecuencias bajas
Para solo conservar las fecuencias bajas, se utilizo un filtro gaussiano, es decir todas las frecuencias fueron multiplicadas por $e^{-\alpha f^2}$ donde $\alpha$ es un valor que se ajustó de manera que se obtuvieran los mejores resultados posibles.
##Filtros Frecuencias altas
Para solo conservar las fecuencias bajas, se utilizo un filtro gaussiano, es decir todas las frecuencias fueron multiplicadas por $\left(1-e^{-\alpha f^2}\right)$ donde $\alpha$ es un valor que se ajustó de manera que se obtuvieran los mejores resultados posibles.

```python
%pylab inline
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
```

#Analisis Imagen 1 
colesterol-1.tif

```python
Im1 = pylab.imread("colesterol-1.tif")
imshow(Im1, cmap = cm.Greys_r)
plt.title('Imagen Original', fontsize=12)

print shape(Im1)
```

```python
T1= fft2(Im1)
Tr1= fft2(Im1)
imshow(log(abs(fftshift(Tr1))))
colorbar()
```

##FiltroBajas
Solo se dejan pasar las frecuencias bajas

```python
low_freq = fftshift(Tr1)
ic=int(Tr1.shape[0])/2
jc=int(Tr1.shape[1])/2
a=5E-3
for i in range(Tr1.shape[0]):
    for j in range(Tr1.shape[1]):
        low_freq[i,j]*=exp(-a*((i-ic)**2 + (j-jc)**2))
```

```python
im1_low = ifft2(fftshift(low_freq))
imshow(abs(im1_low),  cmap=matplotlib.cm.Greys_r)
```

##FiltroAltas
Solo se dejan pasar las frecuencias altas

```python
high_freq = fftshift(Tr1)
ic=int(Tr1.shape[0])/2
jc=int(Tr1.shape[1])/2
a=5E-3
for i in range(Tr1.shape[0]):
    for j in range(Tr1.shape[1]):
        high_freq[i,j]*=(1-exp(-a*((i-ic)**2 + (j-jc)**2)))
```

```python
im1_high= ifft2(fftshift(high_freq))
imshow(abs(im1_high),  cmap=matplotlib.cm.Greys_r)
```

#Análisis imágen 4
full_moon.jpg
Esta imágene es un poco diferente porque tiene tres capas (R,G,B). Luego para realizar este análisis, se redujo una capa utilizando la ecuacion de intensidad relativa para una imágen en RGB. esta es
$$
I=0.2126\times R+ 0.7152\times G+ 0.0722\times B
$$
Posteriormente se aplico el filtro de la misma manera que las otras imágenes

```python
img = plt.imread('full_moon.jpg')
plt.imshow(img)

print shape(img)
```

```python
intensity=zeros(img.shape[0:2])
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        intensity[i,j]=0.2126*img[i,j,0]+ 0.7152*img[i,j,1]+ 0.0722*img[i,j,2]
```

```python
imshow(intensity,cmap=matplotlib.cm.Greys_r)
```

Transformada realizada para cada una de las capas. 

```python
Tr3= fft2(intensity)
imshow(log(abs(fftshift(Tr3))))
```

##Filtro Bajas 
Solo se dejan pasar frecuencias altas

```python
low_freq3 = fftshift(Tr3)
ic=int(Tr3.shape[0])/2
jc=int(Tr3.shape[1])/2
a=5E-3
for i in range(Tr3.shape[0]):
    for j in range(Tr3.shape[1]):
        low_freq3[i,j]*=exp(-a*((i-ic)**2 + (j-jc)**2))
```

```python
im3_low = ifft2(fftshift(low_freq3,axes=(0,1)),axes=(0,1))
imshow(abs(im3_low), cmap=matplotlib.cm.Greys_r)
```

##FiltroAltas 

Solo se dejan pasar las frecuencias altas

```python
high_freq3 = fftshift(Tr3)
ic=int(Tr3.shape[0])/2
jc=int(Tr3.shape[1])/2
a=5E-3
for i in range(Tr3.shape[0]):
    for j in range(Tr3.shape[1]):
        high_freq3[i,j]*=(1-exp(-a*((i-ic)**2 + (j-jc)**2)))
```

```python
im3_high= ifft2(fftshift(high_freq3,axes=(0,1)),axes=(0,1))
imshow(abs(im3_high),cmap=matplotlib.cm.Greys_r)
```

#Análisis
En general lo observado en este trabajo fue que cuando se conservan las frecuencias bajas se conservan los detalles más sobresalientes de la imágen y la imágen se ve más borrosa. Mientras que cuando se conservaban las frecuencias altas solo se conservaban los detalles donde habían cambios abruptos de las imágenes, como justo en los bordes.
