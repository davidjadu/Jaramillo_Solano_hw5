```python
%pylab inline
from scipy.fftpack import fftfreq, fft
```

```python
data = np.loadtxt('circulos.dat')
x=data[:,0]
y=data[:,1]
c=x+1j*y
print len(c)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

ax.scatter(x, y, alpha=0.5, s=1)
```

```python
correlation = fft(x)*fft(y)
print abs(correlation).max()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

ax.plot(abs(correlation[:]))
```

```python
correlation[correlation<2E6]=0
correlation[:1000]=0
```

```python
plot(abs(correlation))
```

```python
clean_data = ifft(correlation)
```

```python
scatter(clean_data.real,clean_data.imag, alpha = 0.5, color= 'r', s=1)
```

```python
scatter(range(len(clean_data)),abs(clean_data), alpha =0.5, s=0.2, color ='r')
print abs(clean_data).max(), abs(clean_data).min()
```

#Plot one circle

```python
R=2
a_1=0
a_2=0.5
b_1=0
b_2=0.5
points=100
theta = linspace(0, 2*pi, points)
```

```python
x_1=R*cos(theta)+a_1
y_1=R*sin(theta)+b_1
x_2=R*cos(theta)+a_2
y_2=R*sin(theta)+b_2
x_random=2.0*R*random.random(points)-R
y_random=2.0*R*random.random(points)-R
```

```python
scatter(x_1,y_1)
scatter(x_2,y_2)
scatter(x_random,y_random)
```

```python
x=concatenate((x_1,x_random))
y=concatenate((y_1,y_random))
scatter(x,y)
points = len(x)
```

```python
correlation = fft( ( abs(fft(x**2+y**2)) )**2 )
```

```python
plot(abs(correlation[:]))
```

```python
correlation_fourier=fft(correlation)

plot(abs(correlation_fourier[1:-11]))
```

```python
correlation_fourier[correlation_fourier<250000]=0
plot(abs(correlation_fourier[5:-1]))
```

```python
correlation_clear=ifft(correlation_fourier)
plot(abs(correlation_clear))
```

```python
correlation_fourier[:5]=0
correlation_fourier
plot(abs(correlation_fourier))
```

```python
correlation_clear=ifft(correlation_fourier)
plot(abs(correlation_clear))
```

```python
radius_squared= ifft( sqrt( ifft(correlation_clear ) ) ) 
```

```python
plot(abs(radius_squared))
radius = sqrt(sum(abs(radius_squared))/len(radius_squared))
```
